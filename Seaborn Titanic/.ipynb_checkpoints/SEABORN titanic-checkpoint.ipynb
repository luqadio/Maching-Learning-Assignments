```python
import pandas as pd, numpy as np, seaborn as sns
```

```python
titanic_train = pd.read_csv('train.csv', delimiter = (','))
titanic_train
```

### 2.Plot the count distribution (Bernoulli) of survival (sns.countplot)

```python
sns.countplot(x= 'Survived', data = titanic_train, hue = "Pclass")
```

### 3. Seperate this distributions now also by Gender (sns.catplot). Experiment with kind and hue arguments 

```python
sns.catplot(x = 'Sex', y = 'Survived', data = titanic_train, kind = 'violin', hue = 'Pclass' )
```

### 4. Plot SibSp vs Survived per gender and Pclass using sns.catplot

```python
sns.catplot(x = 'Survived', y = 'SibSp', hue = 'Pclass', kind = 'bar',
            col = 'Sex', data = titanic_train)
```

### 5. Plot distribution of fare for each Pclass, color by Survived, use sns.boxplot 

```python
sns.boxplot(x = 'Pclass', y = 'Fare', data = titanic_train, hue = 'Survived')
```

## Contingency Table (Crosstab) and Frequency Table
1. Contingency Table (Crosstab) is a type of table that displays how many item belong a class (frequency distribution of a the variables) in matrix format.
2. Frequency tables show categorical variables.

```python
survived_class = pd.crosstab(index = titanic_train['Survived'], columns = titanic_train['Pclass'], margins = True)
survived_class
```

```python
pd.crosstab(index = titanic_train['Survived'], 
            columns = [titanic_train['Pclass'], titanic_train['Sex']], margins = True)
```
